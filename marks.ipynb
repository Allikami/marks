```python
import numpy as np #arrays and math
import pandas as pd #data
import seaborn as sns # graphics
from scipy import stats #statistics
import matplotlib.pyplot as plt #graphics
```

```python
dativ = pd.read_csv('lop.csv')
```

```python
dativ.head()
```

```python
kr_3 = dativ[['k3_final']]
```

```python
dativ[['k4_final', 'last_name']]
```

```python
kr_4 = dativ[['k4_final']]
```

```python
exxam = dativ[['exam']]
```

```python
sns.distplot(kr_3, kde=False)
plt.xlabel('Оценки за контрольную 3')
plt.ylabel('Число оценок')
plt.suptitle('Контрольная номер 3')
plt.show()
```

```python
sns.distplot(kr_4, kde=False)
plt.xlabel('Оценки за контрольную 4')
plt.ylabel('Число оценок')
plt.suptitle('Контрольная номер 4')
plt.show()
```

```python
sns.distplot(exxam, kde=False)
plt.xlabel('Оценки за экзамен')
plt.ylabel('Число оценок')
plt.suptitle('Экзамен')
plt.show()
```

```python
scatter = sns.jointplot(data=dativ, x='k3_final', y='k4_final') # и диаграмму рассеивания и гистограмму показывает
scatter.set_axis_labels('Оценка за контрольную 3', 'Оценка за контрольную 4')
plt.suptitle('Оценки ИП')
```
